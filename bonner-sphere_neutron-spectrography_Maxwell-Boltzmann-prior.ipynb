```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# response function available from the below link:
#https://www.researchgate.net/project/EURADOS-International-comparison-on-neutron-spectra-unfolding-in-Bonner-spheres-spectrometry
#
response_function = pd.read_excel("../BSS_WG06.xlsx")
```

```python
response_function.describe()
```

```python
energy = response_function["E (MeV)"].as_matrix()
num_energy_bins = len(energy)
print energy
print "Number of energy bins = ", num_energy_bins
```

```python
# load the response function of all the 13 Bonner spheres
response_function = response_function.ix[:,"bare":].as_matrix()
response_function.shape
```

```python
# generate Maxwell-Boltzmann spectra, that will be used as prior, from which counts will be extracted
examples = 10000
temperature = np.random.rand(examples) * 1e2
maxwell_spectrum = np.zeros((examples, len(energy)))

for index, temp in enumerate(temperature):
    maxwell_spectrum[index] = np.sqrt(2/np.pi)*energy**2 * np.exp(-energy**2/(2*temp**2)) / temp**3
maxwell_spectrum
```

```python
rand_index = np.random.randint(0,examples)
plt.plot(energy, maxwell_spectrum[rand_index])
plt.xlabel("Energy (MeV)")
plt.ylabel("Flux or count")
plt.title("T = "+str(int(temperature[rand_index])))
```

```python
for i in range(8):
    temp = 440 + i +1
    rand_index = np.random.randint(0,examples)
    #
    ax = plt.subplot(temp)
    plt.plot(energy, maxwell_spectrum[rand_index])
    plt.title("T = "+str(int(temperature[rand_index])))
plt.tight_layout()
```

```python
# convolute spectrum with response function to get counts from all the 13 Bonenr spheres
# counts are the features that will be used for training
counts = np.matmul(maxwell_spectrum, response_function)
counts
```

```python
counts.shape
```

```python
E_bin = np.random.randint(0,num_energy_bins) # choose one energy bin from the spectrum, which will be the target
spectrum_at_energy = np.reshape(maxwell_spectrum[:,E_bin], (examples,1))
spectrum_at_energy.shape
```

```python
# merge features and target
data = np.hstack((counts, spectrum_at_energy))
data
```

```python
# transform data to log space as the values cover several decades
from sklearn.preprocessing import FunctionTransformer
transformer = FunctionTransformer(np.log10)
data = transformer.transform(data)
data
```

```python
# split data into train and test sets
from sklearn.cross_validation import train_test_split
train, test = train_test_split(data, train_size = 0.8)
```

```python
train.shape
```

```python
test.shape
```

```python
X_train = train[:,0:-1]
Y_train = train[:,-1]
X_test = test[:,0:-1]
Y_test = test[:,-1]
```

```python
# standardizing to 0 mean and var = 1
from sklearn.preprocessing import StandardScaler  
scaler = StandardScaler()  
scaler.fit(X_train)  
X_train = scaler.transform(X_train)  
X_test = scaler.transform(X_test)  
#
Y_scaler = StandardScaler()  
Y_scaler.fit(Y_train)  
Y_train = Y_scaler.transform(Y_train)  
Y_test = Y_scaler.transform(Y_test)  
```

```python
X_train[np.random.randint(0,len(X_train))]
```

```python
Y_train[np.random.randint(0,len(Y_train))]
```

```python
# training
from sklearn.neural_network import MLPRegressor
mlp = MLPRegressor()
mlp.fit(X_train, Y_train)
print("Training set score: %f" % mlp.score(X_train, Y_train))
print("Test set score: %f" % mlp.score(X_test, Y_test))
```

```python
Y_pred = mlp.predict(X_test)
plt.plot(Y_scaler.inverse_transform(Y_pred) / Y_scaler.inverse_transform(Y_test))
plt.title("Ratio of prediction to ground truth for test samples")
```
